```python
# All Sofas - testing
allSofas = requests.get('https://www.ikea.com/au/en/catalog/categories/departments/living_room/39130/').text
sofaSoup = bs(allSofas, 'html.parser')
print(sofaSoup.prettify())
```

```python
sofas = sofaSoup.find_all('div', class_="product")

sofaList = []

for s in sofas:
    Title = s.find('span',class_='productTitle').text.strip()
    Desp = s.find('span',class_='productDesp').text.strip()
    spanPrice = s.find('span',class_='regularPrice').text.strip()
    Price = spanPrice.split("Unit")[0].strip()
#     print("-------")
#     print(Title)
#     print(Desp)
#     print(Price)
    
    sofa = {
    'Title': Title,
    # to strip dollar sign and turn data into numbers
    'Price': (Price.replace('$','')),
    'Description': Desp
    }
    sofaList.append(sofa)
```

```python
sofaDF=pd.DataFrame(sofaList)
```
